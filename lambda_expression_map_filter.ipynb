#### data processing functions
- map () - same operation applied on all elements
- filter () - select elements based on condition

```python
map(int, ['1','3','5','10'])
```

```python
# simple extraction from map using loop
for num in map(int, ['1','3','5','10']):
    print(num)
```

```python
for ascii_code in map(ord, ['A','b','C','d']):
    print(ascii_code, end=' ')
```

```python
list(map(ord, ['A','b','C','d']))
```

```python
animals = ['Horse','Rabbit','Deer','Antelope',
           'Giraffe','Lion','Tiger','Wolf',
           'Kangaroo','Meercat','Wolverine']

out = list(map(lambda e: e[0], animals))
print(out)
```

```python
list(map(lambda e: e[:3], animals))
```

```python
fruits = ['apple','apricot','banana','berry',
          'cherry','chestnut','dragonfruit','date',
          'fig','gauva','grape']
print(len(fruits) == len(animals))
```

```python
# first 3 letter of animal and last 3 of fruit
list(map(lambda a,f: a[:3] + f[-3:], animals, fruits))
```

```python
x, y = [1,2,3] , [3,4,5]
z = list(map(lambda i,j: i+j, x, y))
print(x, y, z, sep='\n')
```

```python
# filter
x = [1,-2,-3,4,-5,6]
ans = filter(lambda i: i>0, x)
list(ans)
```

```python
list(filter(lambda name: 'a' in name, animals))
```

```python
# extract names that dont have small letter 'a' in them
list(filter(lambda name: 'a' not in name, animals))
```

```python
# extract animal names with names > 5 chars
list(filter(lambda e: len(e) > 5, animals))
```

```python
class Student:
    def __init__(self, name, marks=[]):
        self.name = name
        self.marks = marks
    
    def total_marks(self): 
        return sum(self.marks)
    
    def __gt__(self, other): 
        if isinstance(other, Student):
            return self.total_marks() > other.total_marks()
        elif isinstance(other, int):
            return self.total_marks() > other
        
    def __repr__(self): 
        return f'Student({self.name},{self.total_marks()})'
    
# outside the class
def is_passed(marks):
    return marks > 200
students = [
    Student('kelsier', [23, 56, 67, 20, 10]),
    Student('Vin', [43, 86, 77]),
    Student('Rashek', [83, 86, 87]),
    Student('Sazed', [53, 86, 67]),
]

list(filter(is_passed, students))
```
