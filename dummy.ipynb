```python
import numpy as np

# Generate motivation levels for a 7-day cycle (1â€“10 scale)
motivation = np.random.randint(1, 11, size=7)
print("Weekly Motivation Matrix:", motivation)

# Core KPIs
avg_motivation = np.mean(motivation)
std_motivation = np.std(motivation)
max_day = np.argmax(motivation) + 1     # +1 for human-readable indexing
min_day = np.argmin(motivation) + 1

print("Average Motivation:", avg_motivation)
print("Standard Deviation:", std_motivation)
print("Peak Performance Day:", max_day)
print("Low-Energy Day:", min_day)
```

```python
# Forecasting next-week momentum with randomized delta noise
noise = np.random.uniform(-1, 1, size=7)
forecast = np.clip(avg_motivation + noise, 1, 10)
print("Next-Week Forecast:", forecast)
```

```python
# Segmentation: crisis and peak days
print(motivation)
crisis_days = motivation < 4
peak_days = motivation > 8

print("Crisis Days (<4):", np.where(crisis_days)[0] + 1)  # +1 for human-readable indexing
print("Peak Days (>8):", np.where(peak_days)[0] + 1)  # +1 for human-readable indexing

# Survival vs Thriving metrics
barely_surviving = np.sum(motivation < 5)
thriving = np.sum(motivation > 7)

print("Barely Surviving Days (<5):", barely_surviving)
print("Thriving Days (>7):", thriving)
```
