```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The emotional journey of learning Python
plt.rcParams['font.family'] = 'Segoe UI Emoji'

events = ['start', 'Hello\nWorld', 'Datatypes', 'Control Flow', 'Data structures',
          'functions', 'OOP', 'Data Analysis', 'Web Dev', 'Projects']
weeks = np.arange(len(events))
confidence = [7, 10, 9, 6, 4, 3, 2, 5, 6, 8]

plt.figure(figsize=(12, 6))
plt.plot(weeks, confidence, marker='s', linewidth=2, markersize=8,
         color='#2962FF', markerfacecolor='orange',)

# Annotate key moments
plt.annotate('Everest', xy=(1, 10), xytext=(1, 4), ha='center',
             arrowprops=dict(facecolor='green', shrink=0.05, width=4))
plt.annotate('Valley of Despair', xy=(6, 2.2), xytext=(6, 5), 
             arrowprops=dict(facecolor='red', shrink=0.05, width=4),
             ha='center')

# 2 vertical bars with hatched pattern indicating challenging phases
plt.axvspan(3.5, 5.5, color='red', alpha=0.1, hatch='o', label='Challenging Phase')
plt.axvspan(6.5, 7.5, color='red', alpha=0.1, hatch='//')

# yline for average confidence
plt.axhline(y=np.mean(confidence), color='blue', label='Average Confidence', alpha=0.1, linewidth=20)
plt.text(-0.4, np.mean(confidence)-.1, 'Average Confidence', color='grey')

# Customize plot
plt.xticks(weeks, events, rotation=45)
plt.xlabel('Learning Milestones')
plt.ylabel('Confidence Level (1-10)')
plt.title('The Emotional Journey of Learning Python')
plt.grid(True, alpha=0.2)
plt.ylim(0, 11)
plt.tight_layout()
# change font style
plt.legend()
plt.show()
```
